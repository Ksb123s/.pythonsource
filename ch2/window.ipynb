```python
w = P.getWindowsWithTitle("제목 없음")[0]

if not w.isActive:
    w.activate()

if not w.isMaximized:
    w.maximize()

P.sleep(2)
# if not w.isMinimized:
#     w.minimize()

w.restore() # 원래 크기로 
```

```python
w = P.getWindowsWithTitle("제목 없음")[0]

if not w.isActive:
    w.activate()

P.write("write")
P.write("pyautogui", interval=0.25)
# P.write("안녕하세요", interval=0.25) 한글 불가능
pc.copy("안녕하세요")
P.hotkey("ctrl","v")
```

```python
def word_write(text):
    pc.copy(text)
    P.hotkey("ctrl","v")
```

```python
w = P.getWindowsWithTitle("제목 없음")[0]

if not w.isActive:
    w.activate()

word_write("write : 반갑습니다.")
```
