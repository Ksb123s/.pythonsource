```python
with open("./resource/score.txt", "r") as f:
    score = f.readline()
    total, avg, count = 0,0,0
    while score:
        total += int(score)
        count +=1
        score = f.readline()
    avg =  total/count 
    print("점수의 합 : {} , 평균점수: {}".format(total, avg))
```

```python
score = []
with open("./resource/score.txt", "r") as f:
    for num in f:
        score.append(int(num))

print(sum(score))
print(sum(score)/ len(score))
```

```python
# score .txt 읽어서 총합과 평균을 result.txt 에 작성
#  총합은 :520
#  평균은 : 86.67
score = []
with open("./resource/score.txt", "r") as f:
    for num in f:
        score.append(int(num))


with open("./resource/result.txt", "w", encoding="utf-8") as f:
    f.writelines(["총합 : %d \n" %int(sum(score)), "평균 : %2.2f" % float(sum(score)/ len(score))])
```

```python
# bmi 지수 측정
#  bmi = weight /(height / 100**2)
# bmi >= 25 과체중 bmi >= 18.5 정상 체중

with open("./resource/info.txt", "r", encoding="utf-8") as f:
    for row in f:
        name, height ,weight = row.strip().split(", ")
        bmi = int(weight)/ ((int(height)/100)**2)
        result = None
        if 25<= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"
        print("\n".join([f"이름 : {name}",f"키 : {height} ",f"몸무게 : {weight} ",f"BMI : {bmi} ",f"결과 : {result}"]))
        print()
    
    
```
